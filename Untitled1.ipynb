```python
sampled_data=data.sample(n=50,replace=True,random_state=31208)
sampled_data
```

```python
estimate_standared_error=sampled_data["aqi"].std() /np.sqrt(sampled_data.shape[0])
estimate_standared_error
```

```python

mean_sample_means=sampled_data["aqi"].mean()
mean_sample_means
```

```python
# Use the correct argument name for your version of scipy
confidence_level = 0.95
confidence_interval = stats.norm.interval(confidence=confidence_level, loc=mean_sample_means, scale=estimate_standared_error)

(confidence_interval)
```

```python


# Assuming 'data' is your DataFrame and it contains 'state_name' and 'Z_Score' columns
# Filtering the data based on the Z-score condition
filtered_data = data[(data['Z_Score'] > 3) | (data['Z_Score'] < -3)]

# Plotting the filtered data as a bar graph
plt.figure(figsize=(9, 8))
plt.bar(filtered_data['state_name'], filtered_data['Z_Score'], color='grey')

# Adding titles and labels
plt.title('Z-Scores Greater than 3 or Less than -3')
plt.xlabel('State Name')
plt.ylabel('Z Score')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.savefig("Z_Score_Filtered_Bar.png")
plt.show()
```
